<div style="border-radius:10px; padding: 15px; background-color:#2c3e50; font-size:115%; text-align:left;width:97%">

Let's create a function that consolidates each model's metrics into a dataframe, facilitating an end-to-end comparison of all models later:



</div>

```python
from sklearn.metrics import classification_report, accuracy_score
import pandas as pd

def evaluate_model(model, X_test, y_test, model_name):
    """
    Evaluates the performance of a trained model on test data using various metrics.
    """

    y_pred = model.predict(X_test)

    # Get classification report
    report = classification_report(y_test, y_pred, output_dict=True)

    # Extracting metrics and ensuring 2 decimal places
    metrics = {
        "precision_0": "{:.2f}".format(round(report["0"]["precision"], 2)),
        "precision_1": "{:.2f}".format(round(report["1"]["precision"], 2)),
        "recall_0": "{:.2f}".format(round(report["0"]["recall"], 2)),
        "recall_1": "{:.2f}".format(round(report["1"]["recall"], 2)),
        "f1_0": "{:.2f}".format(round(report["0"]["f1-score"], 2)),
        "f1_1": "{:.2f}".format(round(report["1"]["f1-score"], 2)),
        "macro_avg_precision": "{:.2f}".format(round(report["macro avg"]["precision"], 2)),
        "macro_avg_recall": "{:.2f}".format(round(report["macro avg"]["recall"], 2)),
        "macro_avg_f1": "{:.2f}".format(round(report["macro avg"]["f1-score"], 2)),
        "accuracy": "{:.2f}".format(round(accuracy_score(y_test, y_pred), 2))
    }

    # Convert dictionary to dataframe
    df = pd.DataFrame(metrics, index=[model_name])

    return df

```

```python
# Define the Gradient Boosting model
gb_model = GradientBoostingClassifier()

# Define the parameter grid for tuning
param_grid = {
    'n_estimators': [200, 300, 400],  # More boosting stages for better accuracy
    'learning_rate': [0.01, 0.05, 0.1],  # Lower learning rate for better generalization
    'max_depth': [4, 5, 6, 7],  # Control tree depth
    'subsample': [0.8, 0.9, 1.0],  # Helps prevent overfitting
    'min_samples_split': [5, 10, 20],  # Reduce overfitting by requiring more samples to split
    'min_samples_leaf': [2, 4, 6]  # Minimum samples required at each leaf node
}

# Use RandomizedSearchCV for faster tuning
random_search = RandomizedSearchCV(
    estimator=gb_model,
    param_distributions=param_grid,
    n_iter=20,  # Increase search iterations for better tuning
    cv=5,  # 5-fold cross-validation for stable results
    n_jobs=-1,
    verbose=2,
    random_state=42
)

# Fit the model to the training data
random_search.fit(X_train, y_train)

# Print the best parameters
print("Best Parameters:", random_search.best_params_)

# Make predictions with the best model
best_gb_model = random_search.best_estimator_
y_pred = best_gb_model.predict(X_test)

```
